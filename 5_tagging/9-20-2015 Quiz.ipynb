# Question 3

```python
from nltk.corpus import brown
from collections import defaultdict

cat_counts = defaultdict()
for category in brown.categories():
    brown_news_tagged = brown.tagged_words(categories=category, tagset='universal')
    tag_fd = nltk.FreqDist(tag for (word, tag) in brown_news_tagged)
    
    pos_counts = defaultdict()
    for x in tag_fd.most_common():
        pos_counts[x[0]] = x[1]
        
    cat_counts[category] = pos_counts
    
df = pd.DataFrame(cat_counts)
df
```

```python
df.apply(lambda x: 100*(x/x.sum()))
```

```python
brown_news_tagged = brown.tagged_words(tagset='universal')
word_tag_pairs = nltk.bigrams(brown_news_tagged)
verb_preceders = [a[1] for (a, b) in word_tag_pairs if b[1] == 'VERB']
fdist = nltk.FreqDist(verb_preceders)
[tag for (tag, _) in fdist.most_common()]
```
